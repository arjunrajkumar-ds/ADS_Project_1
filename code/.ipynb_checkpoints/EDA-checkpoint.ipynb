# Property -

## Numerical -

```python
property_numerical = spark.read.parquet('../clean_data/property_numerical.parquet')
property_numerical = property_numerical.drop('__index_level_0__')
```

```python
property_numerical.limit(3)
```

```python
fig, ax = plt.subplots()
hist(ax, property_numerical.select('gross_sq_metres'), bins = 20, color=['green'])
plt.xlabel('Gross Square Metres (m^2)')
plt.ylabel('Number of Properties')
plt.show()
```

```python
property_numerical = property_numerical.withColumn("log_sq_metres", log10(col("gross_sq_metres")))
fig, ax = plt.subplots()
hist(ax, property_numerical.select('log_sq_metres'), bins = 20, color=['green'])
plt.xlabel('[Log Transformed] Gross Square Metres (m^2)')
plt.ylabel('Number of Buildings')
plt.show()
```

```python
fig, ax = plt.subplots()
hist(ax, property_numerical.select('year_built'), bins = 20, color=['green'])
plt.xlabel('Year Built')
plt.ylabel('Number of Buildings')
plt.show()
```

```python
fig, ax = plt.subplots()
hist(ax, property_numerical.select('sale_price'), bins = 20, color=['green'])
plt.xlabel('Sale Price (AUD)')
plt.ylabel('Number of Buildings')
plt.show()
```

```python
property_numerical = property_numerical.withColumn("log_sale_price", log10(col("sale_price")))
fig, ax = plt.subplots()
hist(ax, property_numerical.select('log_sale_price'), bins = 20, color=['green'])
plt.xlabel('[Log Transformed] Sale Price (AUD)')
plt.ylabel('Number of Buildings')
plt.show()
```
